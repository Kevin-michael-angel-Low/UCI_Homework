```python
# Dependencies
import matplotlib.pyplot as plt
import requests
import pandas as pd
import random as rd
import numpy as np
from scipy.stats import linregress
from citipy import citipy
from api_keys import weather_api_key
```

```python
rd.seed(a = 256)
```

## Generate Cities List

```python
cityList = []

latitude = []
longitude = []

# Create list of randomized latitudes and longitudes
for lat in np.random.randint(-90, 90, 2000):
    latitude.append(lat)
    
for long in np.random.randint(-180, 180, 2000):
    longitude.append(long)
```

```python
cityList = []

for lat, long in zip(latitude, longitude):
    city = citipy.nearest_city(lat, long)
    cityName = city.city_name
    cityList.append(cityName)
```

```python
# Create a list of unique cities, then find the # of cities in that list
uniqueCities = list(dict.fromkeys(cityList))
len(uniqueCities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
uniqueCities
```

```python
# URL information
weatherURL = 'http://api.openweathermap.org/data/2.5/weather?q='
apiKey = '&appid=' + weather_api_key
units = '&units=imperial'

# Create lists to store values found in API
temp = []
humid = []
cloud = []
windSpeed = []
lats = []
longs = []
uCity = []
uCountry = []

counter = 0
setter = 1

for city in uniqueCities:
    
    try:
        response = requests.get(weatherURL + city + apiKey + units).json()
        temp.append(response['main']['temp'])
        humid.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        windSpeed.append(response['wind']['speed'])
        lats.append(response['coord']['lat'])
        longs.append(response['coord']['lon'])
        uCountry.append(response['sys']['country'])
        
        if counter < 49:
            counter += 1
            uCity.append(city)
        else:
            counter = 0
            setter += 1
            uCity.append(city)
        print(f'Processing Record {counter} of Set {setter}, {city}')
    except Exception:
        print('City not found, Skipping')
print("-------------------------\nData Retrieval Complete\n-------------------------")
```

```python
len(uCity)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Add lists to dataframe via dictionary

weatherDict = {
    'City':uCity,
    'Lat': lats,
    'Lng': longs,
    'Max Temp': temp,
    'Humidity': humid,
    'Cloudiness': cloud,
    'Wind Speed': windSpeed,
    'Country': uCountry
}
```

```python
weatherDF = pd.DataFrame(weatherDict)
weatherDF.head()
```

```python
weatherDF.describe()
```

```python
weatherDF.to_csv('output/output.csv')
```

## Linear Regression

```python
# Create new dataframes depending on city location on northern or southern hemispheres
northernDF = weatherDF[weatherDF['Lat'] > 0]
southernDF = weatherDF[weatherDF['Lat'] <= 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create Linear Regression Model
(slope, intercept, rvalue, pvalue, stderr) = linregress(northernDF['Lat'], northernDF['Max Temp'])
regress_values = northernDF['Lat'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))


# Create scatterplot
plt.scatter(northernDF['Lat'], northernDF['Max Temp'])
plt.plot(northernDF['Lat'], regress_values, 'r-')

# Format
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

plt.annotate(line_eq, (3, 30), fontsize = 12, color = 'red')

# Show!
plt.savefig('output/northTemp.png')
print(f'The r-value is: {rvalue**2}')
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create Linear Regression Model
(slope, intercept, rvalue, pvalue, stderr) = linregress(southernDF['Lat'], southernDF['Max Temp'])
regress_values = southernDF['Lat'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))


# Create scatterplot
plt.scatter(southernDF['Lat'], southernDF['Max Temp'])
plt.plot(southernDF['Lat'], regress_values, 'r-')

# Format
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

plt.annotate(line_eq, (-50, 80), fontsize = 12, color = 'red')

# Show!
plt.savefig('output/southTemp.png')
print(f'The r-value is: {rvalue**2}')
plt.show()
```

# Analysis: Maximum Temperature

These graphs show that there is a significant relationship between latitude and the resultant temperature. As location moves towards a more positive latitude (Northern Hemisphere graph), the temprature sharply decreases. Similarly, as location moves from a negative latitude towards 0 (Southern Hemisphere graph), the temperature sharply increases. These two graphs clearly display the relationship between latitude and temperature, with hotter temperatures present towards the equator, and colder temperatures further away from the equator.
